```python
def get_playlist_tracks(playlist_id):

    results = sp.user_playlist_tracks("spotify", playlist_id)
    tracks = results['items']

    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])

    return tracks

tracks = get_playlist_tracks("37i9dQZF1DXcBWIGoYBM5M")
tracks
```

```python
track_ids = [track['track']['uri'] for track in tracks]
track_ids
```

```python
# Getting all the artists and ids

def get_artists_and_ids_from_playlist(playlist_id):

    tracks_from_playlist = get_playlist_tracks(playlist_id)

    artists = []
    artists_ids = []

    for track in tracks_from_playlist:
        artists_info = track['track']['artists']

        for artist_info in artists_info:
            artists.append(artist_info['name'])
            artists_ids.append(artist_info['id'])

    # artists = [(artist["name"], artist["id"]) for track in tracks_from_playlist for artist in track['track']['artists']]

    return list(set(artists)), list(set(artists_ids))

artists, artists_ids = get_artists_and_ids_from_playlist("37i9dQZF1DXcBWIGoYBM5M")
artists
```

```python
# K-Means
from sklearn.preprocessing import StandardScaler

# Select numeric feature columns 
X = data[['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness',
       'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms',
       'time_signature']]


# K-Means is a distance-based algorithm, so we need to scale/normalize the numeric features
X_prep = StandardScaler().fit_transform(X)

# Convert the scaled array back to a DataFrame for easier inspection
X_scaled = pd.DataFrame(X_prep, columns=X.columns)

# Display the first few rows of the scaled features
print(X_scaled.head())
```

```python
# K-Means is a distance based algorithm: we need to scale / normalize:
from sklearn.preprocessing import StandardScaler
X_prep = StandardScaler().fit_transform(X)

pd.DataFrame(X_prep).head()
```

```python
# Clustering
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_prep)
```

```python
# Predicting / assigning the clusters:
clusters = kmeans.predict(X_prep)
clusters
```

```python
# Check the size of the clusters
pd.Series(clusters).value_counts().sort_index()
```

```python
# Explore the cluster assignment in the original dataset
X_df = pd.DataFrame(X)
X_df["cluster"] = clusters
X_df.head()

X_df['cluster'] = X_df['cluster'].apply(lambda x: 0 if x == 1 else 1 if x == 2 else 2)
X_df
```

```python
import numpy as np

# Check the inertia value and see how tightly the clusters are packed
K = range(2, 10)

inertia = []

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=1234)
    kmeans.fit(X_prep)
    inertia.append(kmeans.inertia_)

inertia
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(16,8))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.show()
# 3 clusters might be an optimal choice for your K-means clustering
```

```python
# Using the silhouette score to determine the optimal number of clusters
from sklearn.metrics import silhouette_score
K = range(2, 20)

silhouette = []

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_prep)
    silhouette.append(silhouette_score(X_prep, kmeans.predict(X_prep)))


plt.figure(figsize=(16,8))
plt.plot(K, silhouette, 'bx-')
plt.xlabel('k')
plt.ylabel('silhouette score')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.show()
# The optimal number of clusters is at the highest silhouette score value.
```

```python
# Create a function to automate work
def get_audio_features(artist):
    # get tracks from artist
    results = sp.search(q=f'artist:{artist}', limit=50)
    # extract the track ids
    track_ids = [track['id'] for track in results['tracks']['items']]
    song_names = [track['name'] for track in results['tracks']['items']]
    # extract the audio features
    audio_features = sp.audio_features(track_ids)
    # store audio features in a dataframe
    df = pd.DataFrame(audio_features)
    df['artist'] = artist
    df['song name'] = song_names
    return df
```

```python
# Create a loop and select artists

artists = ['Madonna', 'The rolling stones', 'Backstreet boys']

data = pd.DataFrame()

for artist in artists:
    data_artist = get_audio_features(artist)
    data = pd.concat([data, data_artist])

data = data.reset_index(drop=True)
data
```

```python
x = data[['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']]
# y = data['artist']

scaler = StandardScaler()

x_prep = scaler.fit_transform(x)

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(x_prep)

clusters = kmeans.predict(x_prep)

scaled_data = pd.DataFrame(x_prep, columns=x.columns)
scaled_data['song name'] = data['song name']
scaled_data['artist'] = data['artist']
scaled_data['cluster'] = clusters
scaled_data
```

```python
import pickle
with open("scaler.pkl", "wb") as f:
  pickle.dump(scaler, f)

with open("kmeans.pkl", "wb") as f:
  pickle.dump(kmeans, f)
```

```python
def recommend_song(song, artist):
    cols = ['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']
    # get song id
    results = sp.search(q=f'track:{song} - {artist}', limit=1)
    track_id = results['tracks']['items'][0]['id']
    # get song features with the obtained id
    audio_features = sp.audio_features(track_id)
    # create dataframe
    df_ = pd.DataFrame(audio_features)
    new_features = df_[cols]
    # scale features
    scaled_x = scaler.transform(new_features)
    # predict cluster
    cluster = kmeans.predict(scaled_x)
    # filter dataset to predicted cluster
    filtered_df = scaled_df[scaled_df['cluster'] == cluster[0]]
    filtered_array = np.array(filtered_df[cols], order="C")
    # get closest song from filtered dataset
    closest, _ = pairwise_distances_argmin_min(scaled_x, filtered_array)
    # return it in a readable way
    print('\n[RECOMMENDED SONG]')
    return ' - '.join([scaled_df[scaled_df['cluster'] == cluster[0]].iloc[closest]['song name'].values[0], filtered_df.iloc[closest]['artist'].values[0]])
```
